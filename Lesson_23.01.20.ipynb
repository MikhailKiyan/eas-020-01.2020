```python
from sklearn import datasets

iris = datasets.load_iris()
data = iris.data[:, [0, 3]]
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)
km.fit(data)
km.labels_
```

```python
import matplotlib.pyplot as plt

plt.scatter(data[:, 0][km.labels_ == 0], data[:, 1][km.labels_ == 0], marker='+')
plt.scatter(data[:, 0][km.labels_ == 1], data[:, 1][km.labels_ == 1], marker='o')
plt.scatter(data[:, 0][km.labels_ == 2], data[:, 1][km.labels_ == 2], marker='>')
```

```python
km = KMeans(n_clusters=2)
km.fit(data)
km.labels_
plt.scatter(data[:, 0][km.labels_ == 0], data[:, 1][km.labels_ == 0], marker='+')
plt.scatter(data[:, 0][km.labels_ == 1], data[:, 1][km.labels_ == 1], marker='o')
plt.scatter(data[:, 0][km.labels_ == 2], data[:, 1][km.labels_ == 2], marker='>')
```

```python
km = KMeans(n_clusters=5)
km.fit(data)
km.labels_
plt.scatter(data[:, 0][km.labels_ == 0], data[:, 1][km.labels_ == 0], marker='+')
plt.scatter(data[:, 0][km.labels_ == 1], data[:, 1][km.labels_ == 1], marker='o')
plt.scatter(data[:, 0][km.labels_ == 2], data[:, 1][km.labels_ == 2], marker='>')
plt.scatter(data[:, 0][km.labels_ == 3], data[:, 1][km.labels_ == 3], marker='>')
plt.scatter(data[:, 0][km.labels_ == 4], data[:, 1][km.labels_ == 4], marker='>')
```

```python
ins = []

for i in range(1, 10):
    km = KMeans(n_clusters=i)
    km.fit(data)
    ins.append(km.inertia_)

plt.plot(ins)
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3)
db.fit(data)
db.labels_
```

```python
plt.scatter(data[:, 0][db.labels_ == 0], data[:, 1][db.labels_ == 0], marker='+')
plt.scatter(data[:, 0][db.labels_ == 1], data[:, 1][db.labels_ == 1], marker='o')
plt.scatter(data[:, 0][db.labels_ == 2], data[:, 1][db.labels_ == 2], marker='>')
plt.scatter(data[:, 0][db.labels_ == -1], data[:, 1][db.labels_ == -1], marker='>')
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3, min_samples=3)
db.fit(data)
plt.scatter(data[:, 0][db.labels_ == 0], data[:, 1][db.labels_ == 0], marker='+')
plt.scatter(data[:, 0][db.labels_ == 1], data[:, 1][db.labels_ == 1], marker='o')
plt.scatter(data[:, 0][db.labels_ == 2], data[:, 1][db.labels_ == 2], marker='>')
plt.scatter(data[:, 0][db.labels_ == -1], data[:, 1][db.labels_ == -1], marker='>')
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.35, min_samples=1)
db.fit(data)
plt.scatter(data[:, 0][db.labels_ == 0], data[:, 1][db.labels_ == 0], marker='+')
plt.scatter(data[:, 0][db.labels_ == 1], data[:, 1][db.labels_ == 1], marker='o')
plt.scatter(data[:, 0][db.labels_ == 2], data[:, 1][db.labels_ == 2], marker='>')
plt.scatter(data[:, 0][db.labels_ == -1], data[:, 1][db.labels_ == -1], marker='>')
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.1, min_samples=1)
db.fit(data)
for i in range(-1, max(db.labels_) + 1):
    plt.scatter(data[:, 0][db.labels_ == i], data[:, 1][db.labels_ == i], marker='+')
```
