```python
kn = KNeighborsClassifier(n_neighbors=30)
kn.fit(data[:, [0, 3]], target)
X = data[:, 0]
Y = data[:, 3]
def pdb(X, Y, clf):
    h = .1
    x_min, x_max = X.min() - 1, X.max() + 1
    y_min, y_max = Y.min() - 1, Y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z)
    plt.scatter(X[target==0], Y[target==0], marker='+')
    plt.scatter(X[target==1], Y[target==1], marker='o')
    plt.scatter(X[target==2], Y[target==2], marker='^')
pdb(X, Y, kn)
```

```python
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(data[:, [0, 3]], target)
pdb(X, Y, kn)
```

```python
kn = KNeighborsClassifier(n_neighbors=1)
kn.fit(data[:, [0, 3]], target)
pdb(X, Y, kn)
```

```python
kn = KNeighborsClassifier(n_neighbors=100)
kn.fit(data[:, [0, 3]], target)
pdb(X, Y, kn)
```
